```python
from os import listdir
from os.path import isfile, join, exists
import pandas as pd
import io
import numpy as np
import re
import xlrd
from datetime import datetime
```

```python
def pandas2csv(df,path):
    base_path = "/home/antarctic/Documents/OpenDat/Trabajo/"
    sep = ','
    index = False
    header = True
    encoding = 'latin-1'
    df.to_csv(base_path+path,
              index=index,
              sep=sep,
              header=header,
              encoding=encoding
              )
```
