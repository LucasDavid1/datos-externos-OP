```python
from os import listdir
from os.path import isfile, join, exists
import pandas as pd
import io
import numpy as np
import re
import xlrd
from datetime import datetime
```

```python
def pandas2csv(df,path):
    base_path = "/home/antarctic/Documents/OpenDat/"
    sep = ','
    index = False
    header = True
    encoding = 'latin-1'
    df.to_csv(base_path+path,
              index=index,
              sep=sep,
              header=header,
              encoding=encoding
              )
```

```python
listLines = []

def read_csv_files(folder):
    onlyfiles = [f for f in listdir(folder) if isfile(join(folder, f))]
    onlyfiles = filter(lambda x:x.endswith((".csv")), onlyfiles)
    
    for csv_file in onlyfiles:
        print(csv_file)
        df=pd.read_csv(folder+"/"+csv_file,sep=';',encoding='latin-1',engine='python')
        
        df=df[df.MES_NAC != 0]
        df.loc[(df.DIA_NAC < 10), 'DIA_NAC'] = '0' + df.DIA_NAC.astype(str)
        df.loc[(df.MES_NAC < 10), 'MES_NAC'] = '0' + df.MES_NAC.astype(str)
        df['FECHA_NACIMIENTO'] = df['DIA_NAC'].astype(str)+ '-'+ df['MES_NAC'].astype(str) + '-'+ df['ANO_NAC'].astype(str)
        df['FECHA_NACIMIENTO'] =  pd.to_datetime(df['FECHA_NACIMIENTO'], format='%d-%m-%Y')
        df=df.drop(columns=['DIA_NAC','MES_NAC','ANO_NAC'])
        
        listLines.append(df)
        output = pd.concat(listLines)
        output_path = "Natalidad 3.csv"
        pandas2csv(output,output_path)        
        

read_csv_files('/home/antarctic/Documents/OpenDat/Natalidad')
```
