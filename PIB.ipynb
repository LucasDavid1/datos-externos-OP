```python
p1=pd.read_excel('/home/antarctic/Documents/OpenDat/PIB_por_region/PIB_Region (2006-2010).xls',skiprows=2)
p2=pd.read_excel('/home/antarctic/Documents/OpenDat/PIB_por_region/PIB_Region (2010-2015).xls',skiprows=2)
p3=pd.read_excel('/home/antarctic/Documents/OpenDat/PIB_por_region/PIB_Region (2013-2017).xls',skiprows=2)
```

```python
p1 = p1.rename(columns={'Descripción series': 'REGION'})
p1=pd.melt(p1, id_vars=['REGION'], value_vars=[p1.columns[2],p1.columns[3],p1.columns[4],p1.columns[5],p1.columns[6]],var_name='FECHA', value_name='PIB')
p1=p1[~p1.REGION.str.contains("Subtotal regionalizado")]
p1=p1[~p1.REGION.str.contains("IVA , derechos de importación y otros")]
p1=p1[~p1.REGION.str.contains("Producto Interno Bruto")]
#p1=p1.dropna()

p2 = p2.rename(columns={'Descripción series': 'REGION'})
p2=pd.melt(p2, id_vars=['REGION'], value_vars=[p2.columns[3],p2.columns[4],p2.columns[5],p2.columns[6]],var_name='FECHA', value_name='PIB')
p2=p2[~p2.REGION.str.contains("Subtotal regionalizado")]
p2=p2[~p2.REGION.str.contains("IVA , derechos de importación y otros")]
p2=p2[~p2.REGION.str.contains("Producto Interno Bruto")]
#p2=p2.dropna()

p3 = p3.rename(columns={'Descripción series': 'REGION'})
p3=pd.melt(p3, id_vars=['REGION'], value_vars=[p3.columns[4],p3.columns[5],p3.columns[6]],var_name='FECHA', value_name='PIB')
p3=p3[~p3.REGION.str.contains("Subtotal regionalizado")]
p3=p3[~p3.REGION.str.contains("IVA , derechos de importación y otros")]
p3=p3[~p3.REGION.str.contains("IVA y derechos de importación")]
p3=p3[~p3.REGION.str.contains("Producto Interno Bruto")]
p3['PIB']=p3['PIB']*1000
#p3=p3.dropna()
```

```python
listLines=[]
listLines.append(p1)
listLines.append(p2)
listLines.append(p3)
output = pd.concat(listLines)
output_path = "PIB_por_Region.csv"
pandas2csv(output,output_path)
```
